```python
import selenium
import pandas as pd 
from selenium import webdriver
from selenium.webdriver.common.by import By

from bs4 import BeautifulSoup
import requests
import warnings
warnings.filterwarnings('ignore')

from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException

import time
```

```python
ind_driver = webdriver.Chrome(r"D:\chromedriver.exe")
ind_driver.maximize_window()
#This should open chrome webdriver and maximize the windows
```

```python
#lets load the website
website = input('job site you want to scrape: ')
ind_driver.get(website)  
time.sleep(2)
```

```python
company_name = []
designation = []
date_days = []
job_link = []
company_link = []
```

```python
# Lets create empty lists to store scrapped data

start = 0
end = 50


ind_driver.refresh()
time.sleep(3)

for i in range(start, end): # loop for jumping on next page
    
    job_role = ind_driver.find_elements(By.XPATH, "//a[@class = 'jcs-JobTitle css-jspxzf eu4oa1w0']")
    for j in job_role:
        jobs = j.text
        designation.append(jobs)
        
    companies = ind_driver.find_elements(By.XPATH, "//span[@class = 'companyName']")
    for j in companies:
        comp = j.text
        company_name.append(comp)
        
        
    url = ind_driver.find_elements(By.XPATH, "//a[@class = 'jcs-JobTitle css-jspxzf eu4oa1w0']")
    for j in url:
        link = j.get_attribute('href')
        job_link.append(link)
        
    
    companies_url = ind_driver.find_elements(By.XPATH, "//a[@class = 'turnstileLink companyOverviewLink']")
    for j in companies_url:
        link = j.get_attribute('href')
        company_link.append(link)
        
    date = ind_driver.find_elements(By.XPATH, "//div[@class = 'heading6 tapItem-gutter result-footer']/span[1]")
    for j in date:
            dates = j.text.split('\n')
            date_days.append(dates[1])
        
    time.sleep(2)
    
   
    
    try:#Exception handling if all the pages extracted succefully
        next_page =  ind_driver.find_element(By.XPATH, "/html/body/main/div/div[1]/div/div/div[5]/div[1]/nav/div[6]/a")
        next_page.click()
        time.sleep(2)
    except NoSuchElementException:
        print('All Pages extracted')
        
    
#lets print the length of the lists

print(len(designation))
print(len(company_name))
print(len(date_days))
print(len(job_link))
```

```python
#Lets create a Data Frame so that all the list can be stored at one place
lead_data = pd.DataFrame({'Company Name': company_name, 'Designation': designation, 'Date/Day': date_days, 'Job Advert': job_link})
lead_data
```

```python
#save the data frame into a CSV file.
new_file_name = input('Put the new file name with .csv ext: ')
lead_data.to_csv(new_file_name)
```

```python
#this code is use to fetch the number of review of each company page.
#so that we can sort the big companies with review.

review = []
company_name_direct = []

for i in company_link:
    cmpage = webdriver.Chrome(r'D:\chromedriver.exe')
    cmpage.maximize_window()
    cmpage.get(i)
    
    company = cmpage.find_element(By.XPATH, "//div[@itemprop = 'name']").text
    company_name_direct.append(company)
    
    
    review_d = cmpage.find_element(By.XPATH, "//a[@class = 'css-1opv4v0 emf9s7v0']/div").text
    review.append(review_d)
    
    time.sleep(2) 
#     print(company_name)
#     print(review)
    cmpage.close()
```
